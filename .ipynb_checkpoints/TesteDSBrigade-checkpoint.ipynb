```python
import pandas as pd
import numpy as np
import wordcloud as wc
import matplotlib.pyplot as plt
import seaborn as sns
```

# Etapa 1 - Carregar dados
Começamos carregando os dados e validando que as tabelas estejam corretamente formatadas. Para este fim, foi avaliado o número o tipo de dado em cada tabela, o número de colunas e se estas colunas são as mesmas em todas as tabelas.

```python
df_2016_licitacao = pd.read_csv("2016/licitacao.csv", dtype=object,sep=",")
df_2017_licitacao = pd.read_csv("2017/licitacao.csv", dtype=object,sep=",")
df_2018_licitacao = pd.read_csv("2018/licitacao.csv", dtype=object,sep=",")
df_2019_licitacao = pd.read_csv("2019/licitacao.csv", dtype=object,sep=",")
df_2016_item = pd.read_csv("2016/item.csv", dtype=object,sep=",")
df_2017_item = pd.read_csv("2017/item.csv", dtype=object,sep=",")
df_2018_item = pd.read_csv("2018/item.csv", dtype=object,sep=",")
df_2019_item = pd.read_csv("2019/item.csv", dtype=object,sep=",")
print(df_2016_licitacao.shape)
print(df_2017_licitacao.shape)
print(df_2018_licitacao.shape)
print(df_2019_licitacao.shape)
print(df_2016_item.shape)
print(df_2017_item.shape)
print(df_2018_item.shape)
print(df_2019_item.shape)
```

```python
print(df_2016_licitacao.dtypes)
print(df_2017_licitacao.dtypes)
print(df_2018_licitacao.dtypes)
print(df_2019_licitacao.dtypes)
```

```python
columnList = df_2016_licitacao.columns
for i in range(len(columnList)):
    if (df_2016_licitacao.columns[i] == df_2017_licitacao.columns[i]) & (df_2016_licitacao.columns[i] == df_2018_licitacao.columns[i]) & (df_2016_licitacao.columns[i] == df_2019_licitacao.columns[i]):
        print("all columns are equal.")
    else:
        print("column difference found! ",df_2016_licitacao.columns[i] )
columnList = df_2016_item.columns
for i in range(len(columnList)):
    if (df_2016_item.columns[i] == df_2017_item.columns[i]) & (df_2016_item.columns[i] == df_2018_item.columns[i]) & (df_2016_item.columns[i] == df_2019_item.columns[i]):
        print("all columns are equal.")
    else:
        print("column difference found! ",df_2016_item.columns[i] )
```

O resultado da célula acima prova que todas as colunas em todos os anos apresentam exatamente as mesmas colunas, permitindo criar um nossas tabelas com um simples append.

```python
df_licitacao = df_2016_licitacao.append(df_2017_licitacao,ignore_index=True)
df_licitacao = df_licitacao.append(df_2018_licitacao,ignore_index=True)
df_licitacao = df_licitacao.append(df_2019_licitacao,ignore_index=True)
df_item = df_2016_licitacao.append(df_2017_item,ignore_index=True)
df_item = df_item.append(df_2018_item,ignore_index=True)
df_item = df_item.append(df_2019_item,ignore_index=True)
```

# Etapa 2 - Validação e Preparação de Dados
Antes de rodar qualquer processo de análise, é importante garantirmos a validação das informações. Vamos primeiro cruzar as tabelas e checar quantas ocorrências preenchidas nós temos a nossa disposição. Faremos nossa união usando inner merge para evitar ao máximo ocorrências sem dados.

```python
df_full = df_licitacao.merge(df_item,left_on="ID_LICITACAO",right_on="NR_LICITACAO",how="inner")
```

O erro retornado acima mostra que a documentação enviada apresenta um erro: o campo ID_LICITACAO não existe na tabela "licitacao.csv". Sabemos pela verificação anterior que isso ocorre indiferentemente ao ano. Devemos, portanto, ou encontrar outra coluna que possa ser usada como chava para cruzar as duas tabelas, ou criar uma. A documentação destaca a coluna "NR_LICITACAO", então vamos começar por ela.

```python
df_item.groupby(by="NR_LICITACAO").count()
```

Esse resultado mostra claramente que o número da licitação não é suficiente para resolver isso (em realidade, isso já era perceptível com o comando head() usando anteriormente, mas incluí aqui para mostrar com clareza a impossibilidade deste caminho). "NM_ORGAO" e "ANO_LICITACAO" da licitação também são chaves ineficientes, pois se um órgão realizou mais de uma licitação ele irá apresentar o mesmo problema acima e o ano simplesmente nos retornará a contagem de licitações de cada ano se realizarmos o mesmo teste anterior. A única forma viável, portanto, é criarmos uma ID única para cada licitação.

Para reduzir o tempo de processamento, vamos resumir a tabela às colunas relevantes segundo a documentação fornecida para o teste. Caso novas hipóteses apareçam para ser testadas, mais dados podem ser trazidos posteriormente. Além das colunas para teste, algumas outras colunas serão trazidas com o intuito de criar uma ID melhor para relacionar as tabelas.

```python
columns_licitacao = ["CD_ORGAO","NR_LICITACAO","ANO_LICITACAO","CD_TIPO_MODALIDADE","DS_OBJETO","TP_OBJETO"]
df_lic_filtered = df_licitacao[columns_licitacao]
columns_item = ["CD_ORGAO","NR_LICITACAO","ANO_LICITACAO","CD_TIPO_MODALIDADE","DS_ITEM","VL_UNITARIO_HOMOLOGADO"]
df_item_filtered = df_item[columns_item]
```

```python
df_lic_filtered["ID_LICITACAO"] = df_lic_filtered["CD_ORGAO"].astype(str)+"_"+df_lic_filtered["NR_LICITACAO"].astype(str)+"_"+df_lic_filtered["ANO_LICITACAO"].astype(str)+"_"+df_lic_filtered["CD_TIPO_MODALIDADE"].astype(str)
df_item_filtered["ID_LICITACAO"] = df_item_filtered["CD_ORGAO"].astype(str)+"_"+df_item_filtered["NR_LICITACAO"].astype(str)+"_"+(df_item_filtered["ANO_LICITACAO"]).astype(str)+"_"+(df_item_filtered["CD_TIPO_MODALIDADE"]).astype(str)
```

```python
df_lic_filtered.head()
```

```python
df_item_filtered.head()
```

Como uma licitação pode ter múltiplos itens, a unicidade da chave criada precisa ser confirmada apenas na tabela de licitações. O teste abaixo confirma que nenhuma chave se repete.

```python
aux = df_lic_filtered.groupby(by="ID_LICITACAO").count().sort_values("CD_ORGAO",ascending=False)
```

```python
aux[aux["CD_ORGAO"]>1]
```

Agora sim temos uma ID única para cada ocorrência e podemos realizar o cruzamento das tabelas para prosseguir com as análises. Uma vez que temos todos os dados usados na criação dessa chave em ambas as tabelas, podemos remover as colunas repetidas na "df_item_filtered" antes de realizar o merge. Como agora temos a certeza da unicidade da ID_LICITACAO, podemos realizar esse merge usando um left join. Iremos remover também quaisquer dados em branco para simplificar as análises.

```python
df_item_filtered = df_item_filtered.drop(columns=["CD_ORGAO","NR_LICITACAO","ANO_LICITACAO","CD_TIPO_MODALIDADE"])
df_full = df_lic_filtered.merge(df_item_filtered,on="ID_LICITACAO",how="left").dropna()
```

Outra análise interessante é uma wordcloud com o DS_Item. Uma vez que ele não é padronizado, uma das formas de encontrar quais os itens mais comprados é procurar por termos ou expressões que se repetem com alta frequência, algo facilmente observado usando esse tipo de vizualização. Serão eliminados dessa análise termos como "de" ou "para", uma vez que preposições não agregam nada na nossa análise.

```python
stopwords = ["a","de","da","do","ao","por","pelo","pela","até","para","com","como","em","na","no","nele","e","ou","abnt","o", "x","cm","mm"]

data = " ".join(df_full["DS_ITEM"])
wc_data = wc.WordCloud(width = 3000,
    height = 2000,
    background_color = 'black',
    stopwords=stopwords).generate(data)
# Generate plot
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wc_data, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

Infelizmente, mesmo removendo uma série de termos ainda observamos expressões pouco úteis na nuvem como "mm" ou "validade". Enquanto eles são necessários na descrição completa de um item, eles revelam pouco sobre os dados. Antes de remover esses termos, vale mencionar que no estado atual já podemos observar alguns termos interessantes ocorrendo, como:

- "Fita Adesiva", "Mesa", "embalagem plástica" e outros que sugerem uma alta frequência de compras de materiais de escritório;

- "Serviço" é o termo mais comum, deixando claro que a maior parte das licitações ocorrem para prestação de serviços;

- "Obra", "Madeira", "Aço", "Alvenaria", "concreto" e outros termos sugerem que obras, reparos e manutenção são pautas constantes das licitações. Este resultado é corroborado pela análise de frequência de TP_OBJETO (terceiro mais comum TP_OBJETO = OSE);

- O TP_OBJETO mais comum ("COM" = "Aquisição de alimentos") em combinação com a nuvem nos mostra que "carne bovina" e "farinha" estão entre os itens mais adquiridos (a nuvem sugere também que a carne consumida seja de "primeira qualidade", "1º qualidade" ou "boa qualidade"; todas essas expressões aparecem na nuvem);

- Enquanto os termos "cx" e "caixa" são pouco descritivos, a frequência deles pode ser justificada com os 3 TP_OBJETO mais comuns, que incluem materiais de limpeza, alimentos e materiais para reforma. Esses 3 tipos de ocorrência obviamente incluirão a expressão "caixa" (ou "cx") comumente em seu DS_ITEM, o que nos mostra que a nuvem corrobora os resultados da nossa primeira análise.
